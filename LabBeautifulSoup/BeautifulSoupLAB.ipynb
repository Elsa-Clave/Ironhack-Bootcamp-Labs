```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd

def top_100():
    url = "https://www.billboard.com/charts/hot-100/"
    headers={"User-Agent":"MiAplicacion/1.0"}
    response=requests.get(url, headers=headers)
    soup=BeautifulSoup(response.content, "html.parser")
    
    clean_songs = []
    songs = soup.find_all('h3', class_='a-no-trucate')
    for h3_tag in songs:
        clean = h3_tag.text.strip()
        clean_songs.append(clean)
    
    clean_artists = []
    artists = soup.find_all('span', class_='a-no-trucate')
    for span_tag in artists:
        clean2 = span_tag.text.strip()
        clean_artists.append(clean2)
    
    df = pd.DataFrame({'Song': clean_songs, 'Artist': clean_artists})


    def finder():
        find_song = input('Introduce el nombre de la canción, completo o parcial: ')
        while len(find_song)<3:
            find_song = input('Hay algún error, porfavor, vuelve a introducir el nombre, minimo 3 letras: ')
    
        for i in df['Song']:
            if find_song.lower() in i.lower():
                position = df[df['Song'] == i].index.item()
                song = df['Song'].iloc[position]
                artist = df['Artist'].iloc[position]
                output = input(f'La canción es: {song} de {artist} es correcto?')
                if output.lower() in ['yes','y','si','ok']:
                    return f'Tu canción está en el número {position} del ranking'
                else:
                    pass
        else:
            print('No hemos encontrado la canción')

    print(finder())
    
```
